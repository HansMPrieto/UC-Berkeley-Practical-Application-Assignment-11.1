### Data Preparation

After our initial exploration and fine tuning of the business understanding, it is time to construct our final dataset prior to modeling.  Here, we want to make sure to handle any integrity issues and cleaning, the engineering of new features, any transformations that we believe should happen (scaling, logarithms, normalization, etc.), and general preparation for modeling with `sklearn`. 

Steps:
1) Drop ID, VIN and region columns
2) Drop an entry if the price is 0
3) Drop an entry if year, manufacturer, and model are NaN

```python
# drop Id, VIN, and region columns
data = data.drop(['id', 'region', 'VIN'], axis=1)
```

```python
# drop entry if price is 0
data = data[data['price'] != 0]
```

```python
# drop entry if year, manufacturer, and model are NaN
data = data[data['year'].notna() & data['manufacturer'].notna() & data['model'].notna()]
```

```python
data.head(15)
```

### Modeling

With your (almost?) final dataset in hand, it is now time to build some models.  Here, you should build a number of different regression models with the price as the target.  In building your models, you should explore different parameters and be sure to cross-validate your findings.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# test on Ford's data
ford = data.query("manufacturer == 'ford'")
ford.head()
```

## Linear Regression Model 1

Here, I investigate the relationship between the type of fuel a car has and how it affects its value

```python
# exclude fuel type 'other'
ford_fuels = ford.query("fuel != 'other'")
ford_fuels.head()
```

```python
ford_fuel_features = ['year', 'model', 'fuel']

ford_fuel_features_data = pd.DataFrame(ford_fuels[ford_fuel_features])
ford_fuel_features_data.head()
```

```python
ford_fuel_dummies = pd.get_dummies(ford_fuel_features_data['fuel'])
ford_fuel_model_dummies = pd.get_dummies(ford_fuel_features_data['model'])

ford_fuel_data = pd.concat([ford_fuel_features_data, ford_fuel_dummies, ford_fuel_model_dummies], axis=1)
ford_fuel_data.head()
```

```python
# delete non-numeric columns
del ford_fuel_data['model']
del ford_fuel_data['fuel']

ford_fuel_data.head()
```

```python
# fit the model
ford_fuel_reg = LinearRegression(fit_intercept = False)

ford_fuel_reg.fit(ford_fuel_data, ford_fuels['price'])
```

```python
ford_fuel_data['predictions'] = ford_fuel_reg.predict(ford_fuel_data)
```

```python
ford_fuel_data.head()
```

## Linear Regression Model 3
Here, I investigate the relationship between the condition of a car and how it affects its value

```python
ford_condition_features = ['year', 'model', 'condition']

ford_condition_features_data = pd.DataFrame(ford[ford_condition_features])
ford_condition_features_data.head()
```

```python
ford_condition_dummies = pd.get_dummies(ford_condition_features_data['condition'])
ford_condition_model_dummies = pd.get_dummies(ford_condition_features_data['model'])

ford_condition_data = pd.concat([ford_condition_features_data, ford_condition_dummies, ford_condition_model_dummies], axis=1)
ford_condition_data.head()
```

### Deployment

Now that we've settled on our models and findings, it is time to deliver the information to the client.  You should organize your work as a basic report that details your primary findings.  Keep in mind that your audience is a group of used car dealers interested in fine tuning their inventory.

From the model, and the data itself, we can see that most buyers would prefer to buy Diesel cars, than hybrid, gas, or electric cars.

In addition, it is important to note that the average price of diesel cars is higher, than the average prices for hybrid, gas, or electric cars.

```python
gas_types = ['Diesel', 'Gas', 'Hybrid', 'Electric']
model_means = [ford_fuel_diesel_model_mean, ford_fuel_gas_model_mean, ford_fuel_hybrid_model_mean, ford_fuel_electric_model_mean]
data_means = [ford_fuel_diesel_data_mean, ford_fuel_gas_data_mean, ford_fuel_hybrid_data_mean, ford_fuel_electric_data_mean]
gt_axis = np.arange(len(gas_types))

plt.bar(gt_axis - 0.2, model_means, 0.4, label = 'Mean Price From Model Trained Data')
plt.bar(gt_axis + 0.2, data_means, 0.4, label = 'Mean Price From Original Dataset')
  
plt.xticks(gt_axis, gas_types)
plt.xlabel("Fuel Type")
plt.ylabel("Average Price")
plt.title("Average Ford Car Price Based on Fuel")
plt.legend()
plt.show()
```

According to the line plot below, the prices peak when the odometer reading is around 1000-10000 miles. 

This means that a car will have more value the less it is used. Thus, cars with more than 10000 miles would have a lower value.
